```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
path='https://onepiece.fandom.com/wiki/List_of_Canon_Characters'

response=requests.get(path)
response
soup=BeautifulSoup(response.content,'html.parser')
```

```python
table=soup.find("table")
```

```python
headers = []
for i in table.find_all('th'):
 title = i.text
 headers.append(title)

# table.find_all('th')
```

```python
mydata = pd.DataFrame(columns = headers)
```

```python
for j in table.find_all('tr')[1:]:
 row_data = j.find_all('td')
 row = [i.text for i in row_data]
 length = len(mydata)
 mydata.loc[length] = row
```

```python
mydata.head()
```

```python
mydata=mydata.drop(columns=mydata.columns[0])
```

```python
mydata.columns=['Name','Intro_Chapter','Intro_Episode','Intro_Year','Note']
mydata.head()
```

```python
# i=0
# for name in mydata['Name']:
#     x=name[0:-1]
#     print(x)
for col in mydata.columns:
    mydata[col]=[content[0:-1] for content in mydata[col]]


```

```python
soup1
```

```python
table=soup1.find('table')
```

```python
mydata=pd.DataFrame(columns=['Rank', 'Character','Anime','Manga','Favorites'])
for j in table.find_all('tr')[1:]:
 row_data = j.find_all('td')
 row = [i.text for i in row_data]
 length = len(mydata)
 mydata.loc[length] = row
```

```python
i=0
mydata['ranks']=''
for rank in mydata['Rank']:
    mydata['Rank'][i]=rank.split()[0]
    i+=1
    
```

```python
i=0

for rank in mydata['Character']:
    name=(rank.split('(',1))[0]
    name=str(name)
    mydata['Character'][i]=name[5:-1]
    i+=1
```

```python
i=0
for num in mydata['Favorites']:
#     num=num.astype('str')
    x=(num.split())
    mydata['Favorites'][i]=x[0]
    i+=1
```
